# ***Data Analytics - Lab 1 - Random data visualisation and sampling***

## Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Ex. 2.

```python
a = np.random.uniform(size=(1000,))
b = np.random.normal(size=(1000,))
c = np.random.normal(scale=10, size=(1000,))
d = []

for alpha, beta in [(1, 1), (2, 2), (3, 9)]:
    d.append(np.random.beta(alpha, beta, size=(1000,)))

mu = [0, 0]
cov1 = [[1, 0], [0, 1]]
cov2 = [[2, 1], [1, 2]]
e1 = np.random.multivariate_normal(mu, cov1, (10000,))
e2 = np.random.multivariate_normal(mu, cov2, (10000,))
```

## Ex. 3.

```python
_, ax = plt.subplots(2, 3, figsize=(30, 20))
ax = ax.flatten()
titles = ["Uniform distribution on interval $[0,1]$",
          "Normal distribution: $\mu=0,\sigma=1$",
          "Normal distribution: $\mu=0,\sigma=10$",
          r"Beta distribution: $\alpha=1,\beta=1$",
          r"Beta distribution: $\alpha=2,\beta=2$",
          r"Beta distribution: $\alpha=3,\beta=9$"]

for i, x in enumerate((a, b, c, d[0], d[1], d[2])):
    ax[i].hist(x, bins=15, density=True)
    ax[i].set_title(titles[i])
    ax[i].set_xlabel("$x$")
    ax[i].set_ylabel("$f(x)$")
    ax[i].grid()

plt.show()
```

***The main problem in this exercise was to find the proper number of bins. It has been found empirically. The plots resemble the continous distribution functions quite well. The rule for finding the number of bins $x$ is that it has to be lower than $\sqrt n$, i.e. $x<\sqrt n$, where n is number of samples. However, the number of bins could not be too small because the histograms can become unreadable. In each case the area between the curve is equal to 1, as it should since the curve is corresponding to the density function.***

## Ex. 4.

```python
plt.figure(figsize=(10, 10))
plt.hist2d(e1[:, 0], e1[:, 1], density=True, bins=50)
plt.show()

plt.figure(figsize=(10, 10))
sns.jointplot(e1[:, 0], e1[:, 1], kind='hist')
plt.show()

plt.figure(figsize=(10, 10))
plt.hist2d(e2[:, 0], e2[:, 1], density=True, bins=50)
plt.show()

plt.figure(figsize=(10, 10))
sns.jointplot(e2[:, 0], e2[:, 1], kind='hist')
plt.show()
```

***The plots are quite similar regardless the package used. The rule saying that $x<\sqrt n$ has also been applied here. However, with the too small number of bins histograms are starting to become unreadable.***

## Ex. 5.

```python
alpha_lst = []
beta_lst = []
x_lst = []
y_lst = []

for i in range(1000):
    alpha = np.random.normal(loc=2.0, scale=0.1)
    alpha_lst.append(alpha)
    beta = np.random.normal(loc=0.2, scale=0.2)
    beta_lst.append(beta)
    x = np.random.normal(loc=0.0, scale=3.0)
    x_lst.append(x)
    sigma = np.abs(np.random.normal(loc=0.0, scale=1.0))
    mu = alpha + beta * x
    y = np.random.normal(loc=mu, scale=sigma)
    y_lst.append(y)
```

## Ex. 6.

```python
plt.figure(figsize=(10, 10))
plt.scatter(x_lst, y_lst)
x = np.arange(-6, 6, 0.1)

for i in range(1000):
    alpha = alpha_lst[i]
    beta = beta_lst[i]
    y = alpha + beta * x
    plt.plot(x, y)

plt.title("Randomly generated plot with use of normal distribution")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid()
plt.show()
```

***Since the parameters have been generated with use of normal distribution, the data tend to be similarly located. They are gathered mainly around the center point, as the data in normal distribution gather around the mean.***
